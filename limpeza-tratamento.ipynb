### Lidando com Outliers

---

#### `Identificando outlaiers`

```python
# Identificando atravez de um describe()
df_sem_nulo.describe()
```

```python
# Identificando atravez de um boxplot()
import seaborn as sns

sns.boxplot(x=df_sem_nulo['cliente.tempo_servico'])
```

```python
# selecionando quantis/quartis
Q1 = df_sem_nulo['cliente.tempo_servico'].quantile(.25) # 1° Quartil
Q3 = df_sem_nulo['cliente.tempo_servico'].quantile(.75) # 3° Quartil
IQR = Q3 - Q1
limite_inf = Q1 - 1.5 * IQR
limite_sup = Q3 + 1.5 * IQR
```

```python
# Localizando outliers por index
outliers_index = (df_sem_nulo['cliente.tempo_servico'] < limite_inf) | (df_sem_nulo['cliente.tempo_servico'] > limite_sup)
outliers_index
```

```python
# Aplicando filtro
df_sem_nulo[outliers_index]['cliente.tempo_servico']
```

#### `Substituindo valores para os outliers`

```python
df_sem_out = df_sem_nulo.copy()
```

```python
df_sem_out[outliers_index]['cliente.tempo_servico']
```

```python
# Substituindo outliers pela div de total/mensal para ter o tempo de serviço
df_sem_out.loc[outliers_index, 'cliente.tempo_servico'] = np.ceil(
     df_sem_out.loc[outliers_index, 'conta.cobranca.Total'] /
     df_sem_out.loc[outliers_index, 'conta.cobranca.mensal']
)
```

```python
# Plot para ver resultados
sns.boxplot(x=df_sem_out['cliente.tempo_servico'])
```

```python
df_sem_out[outliers_index][['cliente.tempo_servico','conta.cobranca.mensal', 'conta.cobranca.Total']]
```

#### `Removendo outliers`

```python
df_sem_out[outliers_index]['cliente.tempo_servico']
```

```python
# Fazendo um novo filtro
Q1 = df_sem_out['cliente.tempo_servico'].quantile(.25) # 1° Quartil
Q3 = df_sem_out['cliente.tempo_servico'].quantile(.75) # 3° Quartil
IQR = Q3 - Q1
limite_inf = Q1 - 1.5 * IQR
limite_sup = Q3 + 1.5 * IQR

outliers_index = (df_sem_out['cliente.tempo_servico'] < limite_inf) | (df_sem_out['cliente.tempo_servico'] > limite_sup)

# Aplicando filtros
df_sem_out[outliers_index]
```

```python
df_sem_out = df_sem_out[~outliers_index]
df_sem_out
```

```python
# Verificando Resultados
sns.boxplot(x=df_sem_out['cliente.tempo_servico'])
```

```python
# Resetanto index
df_sem_out.reset_index(drop = True, inplace = True)
df_sem_out
```

```python
# Removeno do coluan ID
df_sem_id = df_sem_out.drop(columns='id_cliente').copy()
df_sem_id
```

```python
# Binarizando para ajudar o time de ML
mapeaento = {
     'nao': 0,
     'sim': 1,
     'masculino': 0,
     'feminino' : 1
}
```

```python
for col in df_sem_id.columns:
     print(f'Coluna: {col}')
     print(df_sem_id[col].unique())
     print('-', 30)
```

```python
colunas = ['telefone.servico_telefone', 'Churn', 'cliente.parceiro', 'cliente.dependentes', 'conta.faturamente_eletronico', 'cliente.genero']
```

```python
df_sem_id[colunas] = df_sem_id[colunas].replace(mapeaento)
df_sem_id
```

#### `One Hot Encoder (dummy)`

```python
# Tratado as variaveis categoricas não binarizadas
df_dummies = pd.get_dummies(df_sem_id, dtype=int).copy()
df_dummies.head()
```
